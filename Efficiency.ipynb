```python
import pandas as pd

df = pd.read_csv('test_data.csv',delimiter = ';')
df
```

```python
closed = df[df['Сделка закрыта'] == 'да']
opened = df[df['Сделка закрыта'] == 'нет']
```

```python
opened
```

```python
import torch
import numpy as np
from sklearn.metrics import accuracy_score



X = closed[["Повторная сделка", "Повторное обращение", "Ответственный","Сумма","Срок закрытия сделки"]].fillna(0)
y = closed["Стадия сделки"]

y_tensor = torch.from_numpy(y.values.reshape(-1, 1)).float()
X_tensor = torch.from_numpy(X.values).float()
X_tensor = (X_tensor - X_tensor.mean(axis=0)) / X_tensor.std(axis=0)
```

```python
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import seaborn as sns

X_train, X_test, y_train, y_test = train_test_split(X_tensor, y_tensor, test_size=0.25, random_state=42)

```

```python
epochs = 10000
learning_rate = 0.001
train_loss_values = []
test_loss_values = []
epoch_count = []

from torch import nn, optim
class LogisticRegressionTorch(nn.Module):
    def __init__(self, input_size, output_size):
        super(LogisticRegressionTorch, self).__init__()
        self.linear = nn.Linear(input_size, output_size)

    def forward(self, X):
        predictions = self.linear(X)
        return torch.sigmoid(predictions)

model = LogisticRegressionTorch(X_train.shape[1], y_train.shape[1])
# определяем функцию потерь — бинарную кросс-энтропию
criterion = nn.BCEWithLogitsLoss()
# определяем алгоритм оптимизации Adam
optimizer = optim.Adam(model.parameters(), lr=learning_rate)
for epoch in range(epochs):
    optimizer.zero_grad()
    predictions = model(X_train)
    loss = criterion(predictions, y_train)
    # вычисляем градиенты
    loss.backward()
    # обновляем параметры
    optimizer.step()
    model.eval()
    with torch.inference_mode():
        # 1. Предсказываем значения на тестовых данных
        test_pred = model(X_test)

        # 2. Вычисляем значения фукнции потерь
        test_loss = criterion(test_pred, y_test.type(torch.float))

        # Выводим результаты
        if epoch % 25 == 0:
            print("Epoch: {:3d} |  Train Loss: {:.3f} |  Test Loss: {:.3f}".format(epoch, loss, test_loss))
            epoch_count.append(epoch)
            train_loss_values.append(loss.detach().numpy())
            test_loss_values.append(test_loss.detach().numpy())

predictions = model(X_train)
predictions =  np.nan_to_num(predictions.detach().numpy())
print('Train accuracy: ',accuracy_score(y_train, predictions.round()))

```

```python
predictions = model(X_test)
predictions =  np.nan_to_num(predictions.detach().numpy())
print('Test accuracy: ',accuracy_score(y_test, predictions.round()))
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(9, 6))

plt.title("Train / test loss")
plt.xlabel("$Epoch$")
plt.ylabel("$Loss$")

plt.plot(epoch_count, train_loss_values)
plt.plot(epoch_count, test_loss_values)

_ = plt.legend(["Train", "Test"])
```
